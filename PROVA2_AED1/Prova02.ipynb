```python
consumo_diario = [50,45,55,60,48,52,47,49,53,51,50,46,54,61,49,53,48,50,52,47,56,50,55,60,48,52,47,49,53,51]
```

```python
def calcular_media_consumo(consumo_diario):
    soma_total = sum(consumo_diario)
    quantidade_dias = len(consumo_diario)
    media = soma_total / quantidade_dias
    return media
media_final = calcular_media_consumo(consumo_diario)
```

```python
print(f"O consumo médio diário foi de: {media_final:.2f} kWh")
```
